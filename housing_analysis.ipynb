```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)

df.head()
```

## Data Exploration
Understanding the dataset structure

```python
df.info()
df.describe()
```

## Price Distribution

```python
plt.hist(df['medv'], bins=20)
plt.title("House Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

## Correlation Analysis

```python
correlation = df.corr()
correlation['medv'].sort_values(ascending=False)
```

## Key Insights

- House prices vary significantly
- Number of rooms strongly impacts price
- Strong potential for predictive modelling
