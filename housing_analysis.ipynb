## Business Problem

Understanding the key factors that influence housing prices can help investors, policymakers, and buyers make better decisions.

This analysis aims to identify which features have the strongest impact on property values.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)

df.head()
```

```python
import seaborn as sns

sns.histplot(df['medv'], bins=30)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

## Price Distribution

```python
plt.hist(df['medv'], bins=20)
plt.title("House Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

## Correlation Analysis

```python
correlation = df.corr()
correlation['medv'].sort_values(ascending=False)
```

## Key Insights

- Properties with more rooms tend to have higher prices  
- Crime rate negatively impacts housing values  
- Socioeconomic factors play an important role in pricing  

This analysis highlights key drivers of the housing market and can support data-driven decision-making.
